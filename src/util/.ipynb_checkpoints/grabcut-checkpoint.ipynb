```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = "/home/tamaru/scene_categorize/main/data/stairs_frames/denkisogokan/denkisogokan_stairs/denkisogokan_stairs_0_img_000010.png"
bgr = cv2.imread(image)
bgr = cv2.cvtColor(bgr, cv2.COLOR_RGB2BGR)
plt.imshow(bgr)
```

```python
h, w = bgr.shape[:2]
mask = np.zeros((h,w), dtype = np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect=(1,1,w,h)
cv2.grabCut(bgr, mask, rect, bgdModel, fgdModel, 10, cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
bgr2 = bgr*mask2[:,:,np.newaxis]
```

```python
plt.imshow(bgr)
plt.imshow(bgr2)
```
